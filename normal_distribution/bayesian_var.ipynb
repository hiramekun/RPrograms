## 事前分布

```r
xplot = seq(0, 0.5, length=100)
a = 0.1
b = 0.1
plot(x = xplot, dgamma(xplot, a, b), type="l", xlab = expression(lambda))
x = c(3.1, 5.7, -2.5)
```

```r
update_a = function(a, x) {
    return(a + length(x)/2)
}

update_b = function(b, x) {
    return(b + 1/2 * length(x) * var(x))
}
```

## 事後分布
以下のcellの計算式より、
 - 事後分布の平均は0.06
 - 事後分布の標準偏差は0.05  

```r
a1 = update_a(a, x)
b1 = update_b(b, x)
print(a1/b1)
print(sqrt(a1/b1**2))
```

グラフを描画すると、青が事前分布で赤が事後分布

```r
plot(x = xplot, dgamma(xplot, a, b), type="l",col = "blue", xlab = expression(lambda), ylim = c(0, 12), ylab="")
par(new = TRUE)
plot(x = xplot, dgamma(xplot, a1,b1), type="l", col = "red", xlab = expression(lambda), ylim = c(0, 12), ylab = "")
```
