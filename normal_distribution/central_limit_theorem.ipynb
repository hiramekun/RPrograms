```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize'] = (15.0, 10.0)
plt.subplots_adjust(wspace=0.4, hspace=0.6)
from IPython.core.display import display, HTML 
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
ans = []
Ys = []
for i in range(1, 1000):
    Ys.extend(np.random.chisquare(2, 1000))
    ans.append(np.mean(Ys))
plt.plot(np.linspace(1, 100000, num=999),ans)
```

```python
N = 10000
Z = []
X = np.linspace(1.75, 2.25, 100)
for i in range(N):
    x = np.random.chisquare(2, 10000)
    Z.append(np.mean(x))
    nbins = 20
    if  (i+1)%(N/10)==0 :
        plt.subplot(2, 5,  (i+1)/(N/10))
        plt.hist(Z, nbins, alpha=0.6, normed=True)
        plt.plot(X, norm.pdf(X, 2, np.sqrt(np.var(x)/i)))
print(f'平均: {np.mean(Z)}\n分散: {np.var(Z)}\n')
print('中心極限定理がなりたつとした時')
print(f'平均: {2}\n分散: {np.var(x)/i}')
```

```python
ans2 = []
Z = []
for i in range(1000):
    Y = np.random.standard_t(2, 1000)
    Z.extend(Y)
    ans2.append(np.mean(Z))
plt.plot(np.linspace(0, 1000, 1000),ans2)
```

```python
N = 10000
Z = []
X = np.linspace(-0.25, 0.25, 100)
for i in range(N):
    x = np.random.standard_t(2, 10000)
    Z.append(np.mean(x))
    if  (i+1)%(N/10)==0 :
        plt.subplot(2, 5,  (i+1)/(N/10))
        plt.hist(Z, 20, alpha=0.6, range=(-0.25, 0.25), normed=True)
        plt.plot(X, norm.pdf(X, np.mean(0), np.sqrt(np.var(x)/i)))
        
print(f'平均: {np.mean(Z)}\n分散: {np.var(Z)}\n')
print('中心極限定理がなりたつとした時')
print(f'平均: {0}\n分散: {np.var(x)/i}')
```
