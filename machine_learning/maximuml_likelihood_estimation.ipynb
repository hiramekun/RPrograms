# 最尤推定量

## 漸近正規性
 - 今回は、二項分布で事象が生起する確率を最尤推定で求める。  
 - その際、最尤推定量の特徴である、漸近正規性を確かめてみる。
    - 漸近正規性とは、標本数が無限に存在するとき、その最尤推定量は正規分布になるというもの。

 - 対象とする二項分布は、コインを投げて表の出る回数がn回の確率を考え、表の出る確率pを0.7として、pを最尤推定で求める。

```r
y = c()
x = 1:1000
for (n in 1:1000)  {
    rand = rbinom(100000, 1, 0.7)
    y = append(y, sum(rand==1)/length(rand))
}
hist(y, ylim = c(0, 300))
x = seq(0.69, 0.71, length=100)
curve(dnorm(x, mean=0.7, sd=sd(y)), add=TRUE)
```

実行すると、確かに最尤推定量は「平均が真のパラメータである0.7、分散はパラメータの分散」である正規分布に従っていることが確認できる。

## 不偏性
 - 標本が無限にたくさんあれば、真のパラメータがもとまるというもの。  
    - 確率論では、これを確率収束するという。

```r
x = seq(10, 10000, length.out = 1000)

y = c()
for (n in x) {
  rand = rbinom(n, 1, 0.7)
  y = append(y, sum(rand==1)/n)
}

plot(x, y, type = "l")
```

実行すると、確かに標本数が増えるたびに、真のパラメータである0.7に収束していく様子が確認できる。
