## (1) パラメータを設定し、事前分布/尤度関数/事後分布の図を書く

```r
N = 5
m = 2
a = 2
b = 5
x = seq(0.01, 1.0, len=100)
par(mfrow = c(3,1))

plot(x, dbeta(x, a, b), type="l", main="prior", xlab = expression(mu))
plot(x, dbinom(x = m, size = N, prob = x), type="l", main="likeliihood function", xlab = expression(mu))

plot_dbeta  = function(x, a, b) {
    plot(x, dbeta(x,  a,  b), type="l", main="posterior", xlab = expression(mu))
}
print_info = function(a, b) {
    mean = a/(a+b)
    var = (a * b) / ((a + b)** 2 * (a + b + 1))
    sprintf("平均値: %4.3f, 分散: %4.3f" , mean, var)
}

a2 = m + a
b2 = N - m + b

plot_dbeta(x, a2, b2)
```

## (2) 事前分布と事後分布の期待値

```r
print("事前分布")
print_info(a, b)
print("事後分布")
print_info(a2, b2)
```
