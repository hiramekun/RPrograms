以下の計算を数値積分により求める. 付随してグラフも記載している. 
 \begin{array} { l } { p ( \pi \geq 0.5 | d a t a ) } \\ { p ( \pi \geq 0.8 | d a t a ) } \end{array}

```r
# 事前分布
f.post = function(x) dbeta(x, 5, 2)
    
integrate(f.post, 0.5, 1)
integrate(f.post, 0.8, 1)
```

```r
x = seq(0, 1, 0.01)
data = c(0, 1, 1, 1, 1)
plot_beta = function(x, a, b, col="red") {
    pre = dbeta(x, a, b)
    plot(x, pre, type="l", col=col, ylim=c(0, 5), ylab="")
}
plot_beta(x, 1, 1, col="red")
par(new=T)
plot_beta(x, 5, 2, col="blue")
```

事前分布を$Beta(0.1, 0.1)$に変えた場合の以下の値と, グラフの記載  
\begin{array} { l } { p ( \pi \geq 0.5 | d a t a ) } \\ { p ( \pi \geq 0.8 | d a t a ) } \end{array}

```r
# 事後分布
f.post = function(x) dbeta(x, 4.1, 1.1)
integrate(f.post, 0.5, 1)
integrate(f.post, 0.8, 1)
```

```r
plot_beta(x, 0.1, 0.1)
par(new=T)
plot_beta(x, 4.1, 1.1, col="blue")
```
